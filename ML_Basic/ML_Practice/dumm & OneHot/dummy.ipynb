```python
import pandas as pd
```

```python
df = pd.read_csv('dummy_varia.csv')
df 
```

```python
df = df.drop('indexx',axis='columns')
df 
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
add_dum=pd.concat([df,dummies],axis='columns')
add_dum
```

```python
drop_dum = add_dum.drop(['town','west windsor'],axis='columns')
drop_dum
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
```

```python
X=drop_dum.drop('price',axis='columns')
X
```

```python
y = drop_dum.price
y
```

```python
reg.fit(X,y)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.score(X,y)
```

```python
from sklearn.preprocessing import LabelEncoder
le =LabelEncoder()
```

```python
df.town = le.fit_transform(df.town)
df
```

```python
X = df[['town','area']]
X
```

```python
y = df.price
```

```python
from sklearn.preprocessing import  OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
X = ct.fit_transform(X)
X
```

```python
X = X[:,1:]
X
```

```python
reg.fit(X,y)
```

```python
reg.predict([[1,0,2800]])
```

```python
reg.predict([[0,1,3400]])
```

```python
reg.score(X,y)
```

```python
model.predict([[45000,4,0,1]])
# ------4 yrs mercedez ------------
```
